```
import MeCab
m = MeCab.Tagger("-Ochasen")
m.parse('')
f = open("neko.txt", "r", encoding="utf_8")
verb_words = []
while True:
  line = f.readline()
  if line:
    node = m.parseToNode(line)
    while node:
      if node.feature.split(",")[0] == u"動詞":
        verb_words.append(node.surface)
      node = node.next
  else:
    break

verb_words
```

```
import MeCab
m = MeCab.Tagger("-Ochasen")
m.parse('')
f = open("neko.txt", "r", encoding="utf_8")
words = []
features = []
while True:
  line = f.readline()
  if line:
    node = m.parseToNode(line)
    while node:
      words.append(node.surface)
      features.append(node.feature)
      node = node.next
  else:
    break
aofb_list = []
for i in range(len(words)-3):
  three_words = words[i:i+3]
  three_features = features[i:i+3]
  f1 = three_features[0].split(',')[0] == u"名詞"
  f2 = three_words[1] == u"の"
  f3 = three_features[2].split(',')[0] == u"名詞"
  if f1 and f2 and f3:
    aofb_list.append(three_words)

aofb_list
```
